## خلاصهٔ کلی تغییرات قیمت اینترنت و خدمات مخابراتی در ایران (۱۳۹۴–۱۴۰۴)

در ۱۰ سال گذشته قیمت‌های اینترنت و خدمات مخابراتی در ایران تغییر کرده‌اند، اما روند آن با تورم عمومی کشور یکسان نبوده: در بسیاری از دوره‌ها تورم کلی مصرف‌کننده (CPI) بسیار بالا بوده، اما قیمت‌های مصرف‌کننده برای «اطلاعات و ارتباطات» یا بسته‌های اینترنتی گاهی کمتر و گاهی بیشتر از میانگین رشد کرده‌اند — بسته به سال، سیاست‌های تعرفه‌گذاری اپراتورها، نرخ ارز و تصمیمات نظارتی.

---

### ### مقایسهٔ کمی و نمونه‌ها

- **تورم کلی (CPI):** شاخص قیمت مصرف‌کننده در دههٔ اخیر بارها جهش کرد و نرخ‌های بالای ۳۰–۴۰٪ در سال‌های ۱۴۰۲ تا ۱۴۰۴ ثبت شده است.
- **قیمت بسته‌های اینترنت موبایل/TD-LTE/ADSL:**  
  نمونهٔ گزارش‌شده: یک بستهٔ ۱۰۰ گیگ شب‌ـ‌روز از حدود **۳۰٬۰۰۰ تومان** به **۵۰٬۰۰۰ تومان** افزایش یافت (۱۴۰۲).  
  یعنی برخی سال‌ها رشد قیمت اینترنت هم‌پای یا حتی بالاتر از تورم بوده است.
- **شاخص قیمتی گروه «اطلاعات و ارتباطات»:**  
  در گزارش‌های رسمی، این شاخص در برخی ماه‌ها از سایر گروه‌های خدمات *کمتر* رشد کرده (مثلاً ~۲۴.۹٪).

**جمع‌بندی عددی:**  
بسته‌های اینترنت برخی سال‌ها جهش قیمت داشته‌اند ولی «شاخص گروه ارتباطات» در چند مقطع رشد آهسته‌تر از تورم بوده است.

---

### اصلی‌ترین دستاوردها و تغییرات ساختاری ده سال گذشته

1. **گسترش پوشش شبکه‌های موبایل 4G/4.5G**  
   دسترسی پهن‌باند موبایل افزایش چشمگیر داشته و بخش زیادی از کاربران، اینترنت ثابت را با LTE جایگزین کرده‌اند.

2. **رول‌اوت فیبرنوری (FTTH)**  
   پوشش فیبرخانگی تا پایان ۱۴۰۲ به حدود ۴.۷ میلیون خانوار گزارش شده است.

3. **افزایش رقابت و تنوع بسته‌ها**  
   ظهور MVNOها، ارائه‌دهندگان جدید و تنوع بسته‌ها موجب کاهش نسبی قیمت به‌ازای هر گیگابایت در برخی دوره‌ها شده است.

4. **بهبود سرعت و کیفیت اینترنت**  
   سرعت متوسط کشور طی دههٔ اخیر افزایش یافته و کیفیت خدمات استریم، آموزش آنلاین و… بهبود داشته است.

5. **گسترش اتصال روستایی**  
   هزاران روستای جدید به اینترنت پرسرعت متصل شده‌اند.

---

### علل اصلی تفاوت نرخ رشد قیمت اینترنت با تورم عمومی

- **قیمت‌گذاری تنظیم‌شده و محدودیت‌های تعرفه‌ای**  
- **نرخ ارز و هزینهٔ تجهیزات وارداتی**  
- **رقابت بین اپراتورها**  
- **تغییر الگوی مصرف کاربران (افزایش مصرف داده)**

---

### پیامدها برای خانوار و کسب‌وکار

- در برخی سال‌ها اینترنت نسبت به تورم کلی «کم‌هزینه‌تر» بوده،  
  ولی در برخی سال‌ها بسته‌ها جهش شدید داشته‌اند.
- توسعهٔ فیبر و 4G فرصت‌های اقتصادی جدیدی ایجاد کرده است.

---

### نقاط ضعف و چالش‌ها

2. نبود دادهٔ دقیق و شفاف در سطح ملی  
3. تفاوت کیفیت بین مناطق شهری و روستایی
