# مقایسه تورم اینترنت و تورم عمومی در ایران (خلاصه ساده)

این نوت‌بوک خلاصه‌ای ساده و قابل‌فهم از تفاوت رشد قیمت اینترنت و مخابرات با تورم عمومی در ده سال گذشته ارائه می‌دهد.

## خلاصهٔ کلی
- تورم عمومی بسیار بالا بوده (۳۰ تا ۴۰٪ یا بیشتر در سال‌های اخیر).
- قیمت اینترنت در بعضی سال‌ها کمتر از تورم رشد کرده و در برخی سال‌ها افزایش شدید داشته.
- توسعهٔ 4G، فیبر نوری، و افزایش پوشش روستایی از دستاوردهای مهم بوده‌اند.

## دستاوردهای مهم
1. گسترش 4G/4.5G  
2. توسعهٔ فیبر نوری (FTTH)  
3. بهبود سرعت و کیفیت  
4. پوشش گسترده‌تر روستایی  
5. بسته‌های متنوع و رقابت بیشتر اپراتورها  

## تفاوت نرخ رشد
- شاخص ارتباطات در برخی سال‌ها رشد کمتر از تورم داشته.
- بسته‌های اینترنتی در برخی سال‌ها جهش قیمتی داشته‌اند.
